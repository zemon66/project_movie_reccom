```python
import pandas as pd
import numpy as np
import ast
```

```python
import nltk
```

```python
cast_data = pd.read_csv('credits.csv')
movies_data= pd.read_csv('movies.csv')
```

```python
movies_data= movies_data.merge(cast_data, on='title')
```

```python
movies_data.shape
```

```python
movies_data= movies_data[['movie_id','title','genres','crew','cast','overview','keywords']]
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies_data['genres']= movies_data['genres'].apply(convert)
movies_data['keyword']= movies_data['keywords'].apply(convert)
```

```python
movies_data.head()
```

```python
movies_data.info()
```

```python
def convert2(obj):
  L=[]
  counter =0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter +=1
    else:
         break
    return L
```

```python
movies_data['cast']= movies_data['cast'].apply(convert2)
```

```python
movies_data.head()
```

```python
def fetch_direct(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i ['job'] == 'Director':
        L.append(i['name'])
    return L
```

```python
movies_data['crew'] = movies_data['crew'].apply(fetch_direct)
```

```python
movies_data.head()
```

```python
movies_data['overview'][0]
```

```python
movies_data['overview'] = movies_data['overview'].fillna('').apply(lambda x: x.split())
```

```python
movies_data.head()
```

```python
movies_data['genres']=movies_data['genres'].fillna('').apply(lambda x:[i.replace(" ","") for i in x])
movies_data['keywords']=movies_data['keywords'].fillna('').apply(lambda x:[i.replace(" ","") for i in x])
movies_data['cast']=movies_data['cast'].fillna('').apply(lambda x:[i.replace(" ","") for i in x])
movies_data['crew']=movies_data['crew'].fillna('').apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_data.head()
```

```python
movies_data['tags'] = movies_data['overview']+ movies_data['genres'] +movies_data['keywords']+ movies_data['cast']+ movies_data['crew']
```

```python
movies_data.head()
```

```python
new_data = movies_data[['movie_id','title','tags']]
```

```python
new_data
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_list(text_list):
    stemmed_list = []
    for text in text_list:
        stemmed_words = [ps.stem(word) for word in text.split()]
        stemmed_text = " ".join(stemmed_words)
        stemmed_list.append(stemmed_text)
    return stemmed_list
```

```python
new_data['tags'] = new_data['tags'].apply(stem_list)
```

```python
new_data['tags'] = new_data['tags'].apply(lambda tags: [tag.lower() for tag in tags] if isinstance(tags, list) else tags)
```

```python
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 5000, stop_words='english')
```

```python
new_data['tags'] = new_data['tags'].apply(lambda tags: ' '.join(tags) if isinstance(tags, list) else tags)
```

```python
cv.fit_transform(new_data['tags']).toarray()
```

```python
vectors = cv.fit_transform(new_data['tags']).toarray()
```

```python
vectors[0]
```

```python
feature_names = cv.get_feature_names_out()
len(feature_names)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
print(similarity)
```

```python
similarity.shape
```

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index = new_data[new_data['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1]) [1:6]
    for i in movies_list:
        print(new_data.iloc[i[0]].title)
```

```python
recommend('The Dark Knight')
```
